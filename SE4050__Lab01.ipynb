```python
import numpy as np

array_4x4 = np.random.exponential(scale=1.0, size=(4,4))

print("4x4 Exponential Distribution Array:")
print(array_4x4)
```

```python
# Generate 100,000 samples from different distributions
exponential_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.uniform(low=0, high=4, size=100000)  # Adjusted range
normal_data = np.random.normal(loc=1, scale=0.8, size=100000)  # Adjusted parameters

# Create histogram comparison
plt.figure(figsize=(10, 6))
plt.hist(exponential_data, density=True, bins=50, histtype="step",
         color="red", label="Exponential", alpha=0.8)
plt.hist(uniform_data, density=True, bins=50, histtype="step",
         color="blue", label="Uniform", alpha=0.8)
plt.hist(normal_data, density=True, bins=50, histtype="step",
         color="green", label="Normal", alpha=0.8)

plt.axis([0, 6, 0, 1.2])  # Adjusted for better visualization
plt.legend(loc="upper right")
plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')

ax.set_title("Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
